```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Box Plot
```

```python
# Median, quartile(25%, 50%, 75%), min max * 1.5 IQR, outlier
```

```python
data = np.array([5,7,9,15,20,22,25,30,32,35,37,40,50,55,60,100])
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(y=data)
plt.title("Box Plot")
plt.ylabel("Data Value")
#plt.grid(True)
plt.show()
```

```python
# Medyan ortadaki çizgidir.
# Ortadaki mavi kutu verilerin daha fazla dağıldığı yerleri gösteriyor.
# Kutunun alt tarafı Q1
# Kutunun üst tarafı Q3
# Alt sakal Q1 - 1.5 IQR
# Üst sakal Q3 + 1.5 IQR
# Üstteki yuvarlak outlier çünkü "100"lük bir çıkıntı var
# Sakallar uzun ise o bölgelerde ortalmaaya uzak verilerin fazla olduğu anlamına gelir.
```
