```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
age_list = [10,20,30,30,30,40,50,60,70,75]
```

```python
weight_list = [20,60,80,85,86,87,70,90,95,90]
```

```python
plt.plot(age_list, weight_list, "r") #Â y -> yellow, g -> green,blue -> b, red -> r
plt.title("Age - Weight")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.savefig("test.png")
plt.show()
```

```python
# Numpy
```

```python
np_age_list = np.array(age_list)
np_age_list
```

```python
np_weight_list = np.array(weight_list)
np_weight_list
```

```python
plt.plot(np_age_list, np_weight_list, "b", marker="o")
plt.title("Age - Weight")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.show()
```

```python
numpy_arr1 = np.linspace(0,10,20)
```

```python
numpy_arr1
```

```python
numpy_arr2 = numpy_arr1 ** 3
```

```python
numpy_arr2
```

```python
plt.plot(numpy_arr1, numpy_arr2, "g*-") 
# Sadece * yaparsan data pointleri gÃ¶sterir. 
#Â Sadece - yaparsan grafiÄin default halidir.
# Sadece -- yaparsan grafiÄi kesikli olarak yapar.
# BunlarÄ± birleÅtirebilirin
plt.title("Title")
plt.xlabel("low")
plt.ylabel("high")
plt.show()
```

```python
#Â Ä°kili grafikler
```

```python
plt.subplot(1,2,1)
plt.plot(numpy_arr1, numpy_arr2, "r*-") 
plt.subplot(1,2,2)
plt.plot(numpy_arr2, numpy_arr1, "g--")
plt.show()
```

```python
# Eksenleri deÄiÅtirme
```

```python
my_figure = plt.figure()
figure_axes = my_figure.add_axes([0.1, 0.1, 0.3, 0.3])
# Soldan boÅluk, alttan boÅluk, geniÅlik, yÃ¼kseklik
figure_axes.plot(numpy_arr1, numpy_arr2, "g")
figure_axes.set_xlabel("X Axis")
figure_axes.set_ylabel("Y Axis")
figure_axes.set_title("Graph Title")
plt.show()
```

```python
new_fig = plt.figure(dpi = 100) #Â ÃÃ¶zÃ¼nÃ¼rlÃ¼k artarttÄ±rÄ±labilir.
new_axes = new_fig.add_axes([0.1,0.1,0.9,0.9])
new_axes.plot(numpy_arr1, numpy_arr1 ** 2, label = "numpy array ** 2")
new_axes.plot(numpy_arr1, numpy_arr1 ** 3, label = "numpy array ** 3")
new_axes.legend(loc=2) # loc konumu belirtir. SayÄ± ile kontrol edebilrisin
plt.show()
```
