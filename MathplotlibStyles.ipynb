```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data1 = np.linspace(0,10,20)
```

```python
data1
```

```python
data2 = data1 ** 2
```

```python
plt.subplots()
```

```python
my_fig, my_axes = plt.subplots()
my_axes.plot(data1, data2, color="#1E0703", alpha=0.5)
my_axes.plot(data2, data1, color="#53CA0A", alpha=0.2)
plt.show()
```

```python
(new_fig, new_axes) = plt.subplots()

new_axes.plot(data1, data1 + 2, color="blue", linestyle='-', 
              marker="*", markersize=8, markerfacecolor="red", linewidth=9.0)
new_axes.plot(data1, data1 + 3, color="red", linewidth=4.0, linestyle='--', 
              marker="o", markersize=8, markerfacecolor="yellow")
new_axes.plot(data1, data1 + 4, color="green", linewidth=3.0, linestyle='-.', 
              marker="+", markersize=4, markerfacecolor="purple")

plt.show()
```
