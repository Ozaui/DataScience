```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("data/athlete_events.csv")
df
```

```python
df.head()
```

```python
df.info()
```

```python
plt.scatter("Height", "Weight", data=df)
plt.title("Athlete Weight vs Height")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

# Seaborn

```python
sns.set_style("whitegrid")
sns.scatterplot(x="Height", y="Weight", data= df)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Athlettes Weight - Height")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.scatterplot(x="Height", y="Weight", data= df)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Athlettes Weight - Height")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x="Height", y="Weight", hue="Sex", data= df)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Athlettes Weight - Height")
plt.show()
```

```python
df["Medal"].unique()
```

```python
df["Sex"].unique()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x="Height", y="Weight", hue="Medal", data= df)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Athlettes Weight - Height")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x="Height", y="Weight", hue="Sex", style="Medal", data= df)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Athlettes Weight - Height")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x="Height", y="Weight", hue="Sex", style="Medal", size="Age", data= df)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Athlettes Weight - Height")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.lineplot(x="Height", y="Weight", hue="Sex", style="Medal", size="Age", data=df)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Athlettes Weight - Height")
plt.show()
```

```python
sns.set_style("white")
sns.displot(x="Height", hue="Sex", data=df)
plt.xlabel("Height")
plt.ylabel("Frequency")
plt.title("Athletes Height Distribution")
plt.show()
```

```python
sns.set_style("white")
sns.displot(x="Height", hue="Sex",kind="kde", data=df)
plt.xlabel("Height")
plt.ylabel("Frequency")
plt.title("Athletes Height Distribution")
plt.show()
```

```python
sns.barplot(x="Medal", y="Height", hue="Sex", data=df)
plt.title("Medal by height")
plt.xlabel("Madel Count")
plt.ylabel("Height")
plt.show()
```

```python
sns.catplot(x="Medal", y="Height", hue="Sex", col="Season", data=df)
plt.show()
```

```python
sns.relplot(x="Medal", y="Height", hue="Sex", col="Season", data=df)
plt.show()
```
