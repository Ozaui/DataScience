```python
# Merge
```

```python
df_merge1 = pd.read_csv('data/7-merge_data1.csv')
df_merge2 = pd.read_csv('data/7-merge_data2.csv')
```

```python
df_merge1
```

```python
df_merge2
```

```python
# Merge - outer join
```

```python
df_merge_outher = pd.merge(df_merge1, df_merge2, on="Employee_ID", how="outer")
```

```python
df_merge_outher
```

```python
# Merge- left join
# 1. tabloda varsa alıyor eğer yoksa almıyor 
```

```python
df_merge_left = pd.merge(df_merge1, df_merge2, on="Employee_ID", how='left')
```

```python
df_merge_left
```

```python
df_merge_right = pd.merge(df_merge1, df_merge2, on="Employee_ID", how='right')
```

```python
df_merge_right
```

```python
# Merge inner join ## Ortaklar alır. Kesişim kümesidir.
```

```python
df_merge_inner = pd.merge(df_merge1, df_merge2, on="Employee_ID", how="inner")
```

```python
df_merge_inner
```
