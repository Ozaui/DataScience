```python
def salary_category(salary):
    if salary < 50000:
        return "Low"
    elif 50000 <= salary <= 70000:
        return "Medium"
    else:
        return "High"
```

```python
df["Salary"].apply(salary_category)
```

```python
df["Salary_Category"] = df["Salary"].apply(salary_category)
```

```python
df
```

```python
# Eğer tecrübe 10 yıldan fazla ise performance skoru +1 arttırılmasını istiyorum.
```

```python
def plus_performance_score(experience):
    if experience > 10:
        return 1
    else:
        return 0
```

```python
df["Experience"].apply(plus_performance_score)
```

```python
df["Adjusted"] = df["Experience"].apply(plus_performance_score)
```

```python
df
```

```python
df["New_Score"] = df["Performance_Score"] + df["Adjusted"]
```

```python
df
```

```python
def adjust_new(row):
    if row["Experience"] > 10:
        return row["Performance_Score"] + 1
    else:
        return row["Performance_Score"]
```

```python
df["Adjusted_Score"] = df.apply(adjust_new, axis=1)
```

```python
df
```

```python
df["Formatted_Name"] = df["Name"].apply(lambda x : x.replace("_", " "))
```
